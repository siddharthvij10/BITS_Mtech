```python
# !pip install together
import os
import json
import sys
from together import Together

TOGETHER_API_KEY = <_____________________________________>
client = Together(api_key=TOGETHER_API_KEY)
```

```python
def get_response(topic):
  response = client.chat.completions.create(
      model="meta-llama/Llama-3-8b-chat-hf",
      messages=[{"role": "user", "content": """Can you create a dataset of 20 samples in json format that will be used for fine tuning a large language model? This dataset should follow the format as shown below and should contain a generic comment related to {} in Instruction and a Sarcastic response in output.

  Dataset Format is given below. Please do not give anything verbose apart from JSON in the output. Do not make statements like "here is a json" etc in the beginning. Just give the json response. Make as many samples possible.
  Instruction: <Question>
  Output: <Answer>""".format(topic)}],
  )
  output = (response.choices[0].message.content)
  python_dict = json.loads(output)
  return python_dict
```

```python
finetuning_dataset = []
trending_topics = ["Climate Change", "Global Warming", "COVID-19 Pandemic",
                   "Social Media Impact", "Artificial Intelligence (AI)",
                   "Cryptocurrency", "Blockchain Technology", "Space Exploration",
                   "Mental Health Awareness", "Renewable Energy", "Political Polarization",
                   "Cybersecurity", "Gender Equality", "Racial Equality", "Human Rights",
                   "Globalization", "Economic Inequality", "Education Reform", "Healthcare Systems",
                   "Internet of Things (IoT)", "Quantum Computing", "Autonomous Vehicles",
                   "Genetic Engineering", "Data Privacy", "Fake News",
                   "Pop Culture and Celebrity Gossip", "Video Games and Esports",
                   "Streaming Services (Netflix, etc.)", "Smartphones and Mobile Technology",
                   "E-commerce", "Work-Life Balance", "Remote Work", "Environmental Conservation",
                   "Digital Transformation", "Mental Health in the Workplace", "Startup Culture",
                   "Gig Economy", "Universal Basic Income", "Sustainable Development", "Urbanization",
                   "Aging Population", "Global Health Issues", "Refugee Crisis", "Terrorism and Security",
                   "Nuclear Proliferation", "Space Colonization", "Artificial General Intelligence",
                   "Virtual Reality (VR)", "Augmented Reality (AR)", "Robotics"]

for topic in trending_topics:
  try:
    finetuning_dataset.extend(get_response(topic))
  except:
    print("Error in ", topic)
```

```python
import pandas as pd

df = pd.DataFrame(columns=['Instruction', 'Output'])
for data in finetuning_dataset:
  df.loc[len(df)] = data

df.head()
```

```python
df.to_csv("dataset.csv")
```
