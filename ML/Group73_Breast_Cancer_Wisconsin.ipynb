```python
from matplotlib import pyplot as plt

def correlation_plot(df):
    # correlation analysis for numerical attributes
    corr_matrix = df.corr()

    # Create correlation heatmap
    plt.figure(figsize=(20,20))
    plt.title('Correlation Heatmap of Dataset')
    a = sns.heatmap(corr_matrix, square=True, annot=True, fmt='.2f', linecolor='black')
    a.set_xticklabels(a.get_xticklabels(), rotation=90)
    a.set_yticklabels(a.get_yticklabels(), rotation=0)           
    plt.show()
    
correlation_plot(df_ip)

# Inference:
# Correlation plot shows multiple attributes which are closely correlated with each other (coefficient > 0.9) 
# and hence, need to be removed
```

```python
from sklearn.model_selection import train_test_split

y = df_ip["diagnosis"]
X = df_ip.drop(["diagnosis"], axis=1)
cols = X.columns
```

```python
# Encode target variable since KNN requires 0 and 1 form of target and not alphabetic
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Train test split - 70:30

training_percent = 0.7
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = training_percent,random_state=100)
```

```python
# Standardize the data (X) before feeding to KNN. We need to standardize since KNN is a distance based algo and if we do not
# standardize variables, then the one with bigger range of values will overpower the distance values

from sklearn.preprocessing import StandardScaler

std_scalar = StandardScaler()
std_scalar.fit(X_train)
X_train_std = std_scalar.transform(X_train)
X_test_std = std_scalar.transform(X_test)
X_train_std = pd.DataFrame(X_train_std, columns=cols)
X_test_std = pd.DataFrame(X_test_std, columns=cols)
```

```python
X_train_std.head(10)
```

```python
X_test_std.head(10)
```

```python
# Fitting clasifier to the Training set

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score

def knn(X_train, y_train, X_test, y_test, k):
    print("k = ", k)

    # Instantiate learning model
    classifier = KNeighborsClassifier(n_neighbors=k)

    # Fitting the model
    classifier.fit(X_train, y_train)

    # Predicting the Test set results
    y_pred = classifier.predict(X_test)
    
    cm = confusion_matrix(y_test, y_pred)
    print("confusion matrix is \n", cm)
    
    accuracy = accuracy_score(y_test, y_pred)*100
    print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
    print('\n')
    
    y_pred_proba = classifier.predict_proba(X_test)[:,1]
    return classifier, y_pred, y_pred_proba

k = 5
_, y_pred, _ = knn(X_train_std, y_train, X_test_std,y_test, k)

k = 3
_, y_pred, _ = knn(X_train_std, y_train, X_test_std,y_test, k)
```

```python
# Finding right value of k by using a graph between train and test accuracy. The val of k with highest test freq selected

import numpy as np

#Setup arrays to store training and test accuracies
neighbors = np.arange(1,15)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

def iterate_over_diff_k(X_train, y_train, X_test, y_test, neighbors, train_accuracy, test_accuracy):
    for i,k in enumerate(neighbors):
        #Setup a knn classifier with k neighbors
        classifier, _, _ = knn(X_train, y_train, X_test, y_test, k)    

        #Compute accuracy on the training set
        train_accuracy[i] = classifier.score(X_train, y_train)

        #Compute accuracy on the test set
        test_accuracy[i] = classifier.score(X_test, y_test) 
    
    print("train_accuracy : ", train_accuracy)
    print("test_accuracy : ", test_accuracy)
    
    #Generate plot
    plt.title('k-NN Varying number of neighbors')
    plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
    plt.plot(neighbors, train_accuracy, label='Training accuracy')
    plt.legend()
    plt.xlabel('Number of neighbors')
    plt.ylabel('Accuracy')
    plt.show()

iterate_over_diff_k(X_train_std, y_train, X_test_std,y_test, neighbors, train_accuracy, test_accuracy)

# Inference
# K = 7 has one of the highest test accuracy except k= 13. Hence choose k = 7
# k val 7, 9, 11, 13 have good test accuracy. 11, 13 isnt selected since train accuracy is reduced here than test and 
# this looks like underfitting. Amoung 7 and 9, both look decent in terms of accuracy without any over or underfitting.
# we select 7 since there will be less computation required for k=7 than k=9 in finding neighbours.
```

```python
# Classification report

def class_report(y_test, y_pred):
    print(classification_report(y_test, y_pred))
```

```python
# ROC - AUC curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

def roc_auc(classifier, k, y_pred_proba, y_test):
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    
    plt.plot([0,1],[0,1],'k--')
    plt.plot(fpr,tpr, label='Knn')
    plt.xlabel('fpr')
    plt.ylabel('tpr')
    plt.title('Knn(n_neighbors={}) ROC curve'.format(k))
    plt.show()
    
    auc_val = roc_auc_score(y_test,y_pred_proba)
    print("auc is {}".format(auc_val))
```

```python
# run classifier with selected k = 7 and show classification report and AUC.
from sklearn.metrics import classification_report

k=7
classifier, y_pred, y_pred_proba = knn(X_train_std, y_train, X_test_std,y_test, k)
class_report(y_test, y_pred)
auc_val = roc_auc(classifier, k, y_pred_proba, y_test)

# Inference: Precision, Recall and F1 show very good values and AUC also has a good value of 98. Good classifier.
```

```python
# Let's try k-fold (k=7 as selected earlier) cross validation to find val of F1 score and other eval parameters
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from numpy import mean
from numpy import std
from sklearn.pipeline import Pipeline

# define pipeline
def make_pipeline_for_cv(k):
    steps = list()
    steps.append(('scaler', StandardScaler()))
    steps.append(('model', KNeighborsClassifier(n_neighbors=k)))
    pipeline=Pipeline(steps=steps)
    return pipeline

def cross_val_func(splits, pipeline, k):
    cv = KFold(n_splits=splits,random_state=100, shuffle=True)  
    scores = cross_val_score(pipeline, X, y, cv=cv, scoring='f1')
    scores_acc = cross_val_score(pipeline, X, y, cv=cv, scoring='accuracy')
    scores_roc_auc = cross_val_score(pipeline, X, y, cv=cv, scoring='roc_auc')

    # report performance
    # print(scores)
    mean_score = mean(scores)*100
    mean_acc_score = mean(scores_acc)*100
    mean_auc_score = mean(scores_roc_auc)*100
    print("k value =", k)
    print('F1: %.3f (%.3f)' % (mean_score, std(scores)*100))
    print('ACC: %.3f (%.3f)' % (mean_acc_score, std(scores_acc)*100))
    print('AUC: %.3f (%.3f)' % (mean_auc_score, std(scores_roc_auc)*100))
    return mean_score, mean_acc_score, mean_auc_score
                  
k = 7
pipeline = make_pipeline_for_cv(k)

splits = 4  # n splits is 4 since we selected 70:30 ratio in train test split before and 4 makes it close to this ratio
cross_val_func(splits, pipeline, k)

# Inference:
# Though F1 score came out as 95 in the classification report in previous cell. Here, we see that in cross validation, 
# the score is actually low (92 - mean). This is because the data for one class was grouped together in previous cell
# by default. This behaviour was changed in cross validation because we have set shuffle=True in KFold function. This 
# makes the data shuffle before the folds are performed. This makes sure that target categories are not grouped together
# anywhere in the data.
```

```python
# let's try cross validation with different values of k. This is to validate best value of k (7) selected earlier.

k_list = list(range(1,14,1)) # creating list of K for KNN
cv_scores = [] # creating list of cv scores
cv_scores_acc = []
cv_scores_auc = []

# perform 4-fold cross validation for 70:30 ratio that was used in train test split before. 
# This is to be done for diff vals of k.
for k in k_list:
    pipeline = make_pipeline_for_cv(k)
    mean_score, mean_acc_score, mean_auc_score = cross_val_func(splits, pipeline, k)
    cv_scores.append(mean_score)
    cv_scores_acc.append(mean_acc_score)
    cv_scores_auc.append(mean_auc_score)
    print('\n')
    
def plot_scores_for_diff_k(kpi, scores, k_list):    
    plt.figure()
    plt.figure(figsize=(8,5))
    plt.title('The optimal number of neighbors based on {} score'.format(kpi), fontsize=20, fontweight='bold')
    plt.xlabel('Number of Neighbors K', fontsize=15)
    plt.ylabel('{} score'.format(kpi), fontsize=15)
    sns.set_style("whitegrid")
    plt.plot(k_list, scores)

    plt.show()

plot_scores_for_diff_k("F1", cv_scores, k_list)
plot_scores_for_diff_k("ACC", cv_scores_acc, k_list)
plot_scores_for_diff_k("AUC", cv_scores_auc, k_list)

# Inference: 
# Based on F1 score and ACC score upon cross validation with different values of k, we see that for k=9, we have max value
# of F1 and ACC. The value of k selected earlier was 7 and was not based on cross validation. Best k now moved to 9 since
# the distribution of data was not that proper(same categories were lying together) in earlier dataset. Upon k-fold cross
# validation with shuffle=True, we have shuffled the data and hence, dataset is more scattered with no groups of same
# target categories.
```

```python
# We tried couple of methods to find right value of k.

# 1- We compared graph of test and train accuracy with different value of k. This lead to k=7 as best k.
# 2- We basically tried finding right value of k by using a for loop and then finding eval matrix (accuracy, F1) score 
# using cross validation for each value of k. We then compared the eval matrix values in the graph to find the val of k 
# for which eval matrix val is maximum. This concluded val of k = 9 as best value.

# Conclusion - The modification from k =7 in method 1 to k =9 in second method is because of the reason given in previous cell.
```
